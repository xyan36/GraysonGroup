```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

% matplotlib inline
```

```python
# fname = "210210//210210_Bi2Te3_p4_contact_test2.txt"
# fname2 = "210210//210210_Bi2Te3_p4_contact_test3.txt"
fname = "210222//210222_Bi2Te3_n4_contact_test2.txt"
df = pd.read_csv(fname, sep = ",", header = 0)
#df2 = pd.read_csv(fname2, sep = ",", header = 0)
```

```python
df['Time'] = pd.to_datetime(df['Date Time'])
df['TimeDur'] = df['Time'] - df['Time'][0]
# df2['Time'] = pd.to_datetime(df2['Date Time'])
# df2['TimeDur'] = df2['Time'] - df2['Time'][0]
```

```python
fig, axs = plt.subplots(4,1, figsize = (8,6)) 
mask = pd.Timedelta(5000*1e9, unit = 'ns')
x = (df['TimeDur'][df['TimeDur'] > mask] - mask)*1e-9
axs[0].plot(x, df.R_GaSn[df['TimeDur'] > mask], label = 'GaSn')
axs[0].set_xlabel('time(s)')
axs[0].set_ylabel('R(Ohms)')
axs[0].legend(loc = 'upper right')
axs[0].set_ylim([1.75, 1.85])

axs[1].plot(x, df.R_coldIn[df['TimeDur'] > mask], label = 'Cold In')
axs[1].set_xlabel('time(s)')
axs[1].set_ylabel('R(Ohms)')
axs[1].legend(loc = 'upper right')
axs[1].set_ylim([5, 7])

axs[2].plot(x, df.R_Cerrotru[df['TimeDur'] > mask], label = 'Cerrotru')
axs[2].set_xlabel('time(s)')
axs[2].set_ylabel('R(Ohms)')
axs[2].legend(loc = 'upper right')
axs[2].set_ylim([2.15,2.25])

axs[3].plot(df.TimeDur*1e-9, df.R_hotIn, label = 'Hot In')
axs[3].set_xlabel('time(s)')
axs[3].set_ylabel('R(Ohms)')
axs[3].legend(loc = 'upper right')
axs[3].set_ylim([2.35,2.45])
plt.tight_layout()

fig.savefig(fname[:-5] + '_analysis2.png', dpi = 100)
```
